## Evolset-Centremost get results

```python
from source._resultsGatherer import *
from source._helpersV4 import logSave,load_parameters
```

```python
centremost = 4
diff_type = 'diff'
arch_type = 'pre_adjw'

p_path = 'experiments/shms/shms_p.json'
p = load_parameters(p_path)

# save_filename = 'ECCD_ch{}_'.format(channel)
save_folder = 'results/center_{}/{}'.format(centremost,arch_type)
p['save_folder'] = save_folder
# p['save_filename'] = save_filename

p['n_workers']= 1
p['n_slaves'] = 0
p['gpus']     = []
p['gpu_mem']  = 1000*5


#Creating cma output files container folder
folder = '{}/shms_s{}_{}/{}'.format(p['experiment_path'],centremost,diff_type,save_folder)
if not os.path.exists(folder):
    os.makedirs(folder)
```

```python

l_diff_type = [diff_type]
l_exp = [1]
l_inputs = [10]
l_gen = [str(i).zfill(4) for i in range(10,301,10)]


l_p = []
is_new_dataset = True
for diff_type in l_diff_type:
    if diff_type == 'non-diff':
        l_filter = ['mio']#['none','mio','softmax','none','mio','softmax']
        l_loss_f = ['bin_f1']#['bin_f1','bin_f1','bin_f1','bin_ap','bin_ap','bin_ap']
        opt_direction = 'maximize'
    elif diff_type == 'diff':
        l_filter = ['softmax']#['none','mio','softmax','mio','softmax']
        l_loss_f = ['mul_ce']#['mae','mae','mae','bce','bce']
        opt_direction = 'minimize'
    
    for inputs in l_inputs:
        is_new_inputs = True
        for exp in l_exp:
            for filter,loss_f in zip(l_filter,l_loss_f):
                p_exp = p.copy()
                p_exp['is_new_dataset'] = is_new_dataset
                is_new_dataset = False
                p_exp['is_new_inputs'] = is_new_inputs
                is_new_inputs = False
                p_exp['gen_list'] = l_gen
                p_exp['seed'] = p['available_seeds'][exp-1]
                p_exp['nn_input_size'] = inputs
                p_exp['window_size']   = inputs
                p_exp['output_filter'] = filter
                p_exp['nn_loss_function'] = loss_f
                p_exp['opt_direction'] = opt_direction

                exp_name = 'exp{}_ins{}_{}_{}'.format(exp,inputs,filter,loss_f,arch_type)
                exp_path = '{}shms_s{}_{}/{}/'.format(p_exp['experiment_path'],centremost,diff_type,exp_name)
                p_exp['data_path'] += 's{}_20Hz_TEset_nonNorm_addedMissSqSat.csv'.format(centremost)
                p_exp['arch_fname'] = '{}{}/ind_'.format(exp_path,arch_type)
                p_exp['adjw_fname'] = '{}{}/cma/cma_'.format(exp_path,arch_type)
                p_exp['csv_fname']  = '{}_g'.format(exp_name)
                p_exp['zip_fname']  = '{}shms_s{}_{}/results/{}/{}.zip'.\
                    format(p['experiment_path'],centremost,diff_type,arch_type,exp_name)
                p_exp['exp_print'] = '{} -> centremost {} input size {} filter {}-{} '.\
                    format(exp_name,centremost,inputs,filter,loss_f)
                #updating output size in list of p's
                if 'target_size' in p_exp: # updating nn output size
                    p_exp['nn_output_size'] = len(p_exp['target_size'])
                l_p += [p_exp]
```

```python
 evaluate_agent_save(l_p)
```

```python
l_channels = [4]
l_diff_type = ['diff']
l_exp = [1]
l_seeds  = [123,777,555,10331,7575]
l_inputs = [10]
l_gen = [str(i).zfill(4) for i in range(10,301,10)]


l_p = []
for channel in l_channels:
    is_new_dataset = True
    for diff_type in l_diff_type:
        if diff_type == 'non-diff':
            l_filter = ['mio']#['none','mio','softmax','none','mio','softmax']
            l_loss_f = ['bin_f1']#['bin_f1','bin_f1','bin_f1','bin_ap','bin_ap','bin_ap']
            opt_direction = 'maximize'
        elif diff_type == 'diff':
            l_filter = ['softmax']#['none','mio','softmax','mio','softmax']
            l_loss_f = ['mul_ce']#['mae','mae','mae','bce','bce']
            opt_direction = 'minimize'
        
        for inputs in l_inputs:
            is_new_inputs = True
            for exp in l_exp:
                for filter,loss_f in zip(l_filter,l_loss_f):
                    p_exp = p.copy()
                    p_exp['is_new_dataset'] = is_new_dataset
                    is_new_dataset = False
                    p_exp['is_new_inputs'] = is_new_inputs
                    is_new_inputs = False
                    p_exp['gen_list'] = l_gen
                    p_exp['seed'] = l_seeds[exp-1]
                    p_exp['nn_input_size'] = inputs
                    p_exp['window_size']   = inputs
                    p_exp['output_filter'] = filter
                    p_exp['nn_loss_function'] = loss_f
                    p_exp['opt_direction'] = opt_direction

                    exp_name = 'exp{}_ins{}_{}_{}'.format(exp,inputs,filter,loss_f,arch_type)
                    exp_path = '{}shms_s{}_{}/{}/'.format(p_exp['experiment_path'],channel,diff_type,exp_name)
                    if testing:
                        p_exp['data_path'] += 's{}_20Hz_TEset_nonNorm_addedMissSqSat.csv'.format(channel)
                    else:
                        p_exp['data_path'] += 's{}_20Hz_realset.csv'.format(channel)
                    # p_exp['arch_fname'] = '{}{}/ind_{}.npy'.format(exp_path,arch_type,g)
                    # p_exp['adjw_fname'] = '{}{}/cma/cma_{}'.format(exp_path,arch_type,g)
                    # p_exp['csv_fname'] = '{exp_name}_g{}.csv'.format(g)
                    p_exp['arch_fname'] = '{}{}/ind_'.format(exp_path,arch_type)
                    p_exp['adjw_fname'] = '{}{}/cma/cma_'.format(exp_path,arch_type)
                    p_exp['csv_fname']  = '{}_g'.format(exp_name)
                    p_exp['zip_fname']  = '{}shms_s{}_{}/results/{}/{}.zip'.\
                        format(p['experiment_path'],channel,diff_type,arch_type,exp_name)
                    p_exp['exp_print'] = '{} -> channel {} input size {} filter {}-{} '.\
                        format(exp_name,channel,inputs,filter,loss_f)
                    #updating output size in list of p's
                    if 'target_size' in p_exp: # updating nn output size
                        p_exp['nn_output_size'] = len(p_exp['target_size'])
                    l_p += [p_exp]
```
